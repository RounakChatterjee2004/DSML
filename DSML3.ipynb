```python
import pandas as pd
data=pd.read_csv("dataset/House Data.csv")
```

```python
# Remove leading/trailing spaces
data['GrossSquareMeters'] = data['GrossSquareMeters'].str.strip()

# Check for missing values
print(data["GrossSquareMeters"].isnull().sum())
```

#Standard Deviation And Variance

```python
# Remove 'm2' from GrossSquareMeters and convert to numeric
data['GrossSquareMeters'] = data['GrossSquareMeters'].str.replace('m2', '')
data['GrossSquareMeters'] = pd.to_numeric(data['GrossSquareMeters'])
print(data[['price', 'GrossSquareMeters']].head())
```

```python
#Here in this dataset only NumberFLoorsofBuilding has numeric values so numerical operations like std deviation or variance will only work on this attribute/feature,
#we can convert the non  umeric into numeic but it will fill as NaN values whose std deviation,variance will also lead to NaN

print(data['NumberFloorsofBuilding'].std())
print(data['NumberFloorsofBuilding'].var())
print("\n________________________________\n")
print(data['GrossSquareMeters'].std())
print(data['GrossSquareMeters'].var())
```

#Percentiles

```python
percentiles = data['NumberFloorsofBuilding'].quantile([0.25, 0.50, 0.75])  # 25th, 50th, and 75th percentiles
print(percentiles)
print("\n________________________________\n")
percentiles = data['GrossSquareMeters'].quantile([0.25, 0.50, 0.75])  # 25th, 50th, and 75th percentiles
print(percentiles)
```

#Histogram

```python
import matplotlib.pyplot as plt

data.hist(bins=4,edgecolor="black")
plt.show()
```
