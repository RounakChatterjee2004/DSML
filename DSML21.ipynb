```python
import numpy as np
import pandas as pd


# Features (exclude species column)
X = df.drop("species", axis=1).values  # Only the numerical columns are used for clustering

# Initial centroids (randomly picking from dataset)
m1 = X[0]  # First data point as the initial centroid
m2 = X[50]  # 51st data point as another centroid
m3 = X[100]  # 101st data point as the third centroid
m4 = X[10]  # 11th data point as the fourth centroid (you can change this)

# Function to calculate Euclidean distance
def euclidean_distance(point, centroid):
    return np.sqrt(np.sum((point - centroid) ** 2))

# Perform one iteration of K-means for 4 clusters
def k_means_iteration(X, m1, m2, m3, m4):
    cluster1, cluster2, cluster3, cluster4 = [], [], [], []
    for point in X:
        # Calculate distances to centroids
        dist_to_m1 = euclidean_distance(point, m1)
        dist_to_m2 = euclidean_distance(point, m2)
        dist_to_m3 = euclidean_distance(point, m3)
        dist_to_m4 = euclidean_distance(point, m4)

        # Assign point to the nearest centroid
        if dist_to_m1 < dist_to_m2 and dist_to_m1 < dist_to_m3 and dist_to_m1 < dist_to_m4:
            cluster1.append(point)
        elif dist_to_m2 < dist_to_m1 and dist_to_m2 < dist_to_m3 and dist_to_m2 < dist_to_m4:
            cluster2.append(point)
        elif dist_to_m3 < dist_to_m1 and dist_to_m3 < dist_to_m2 and dist_to_m3 < dist_to_m4:
            cluster3.append(point)
        else:
            cluster4.append(point)

    # Update centroids
    new_m1 = np.mean(cluster1, axis=0) if cluster1 else m1
    new_m2 = np.mean(cluster2, axis=0) if cluster2 else m2
    new_m3 = np.mean(cluster3, axis=0) if cluster3 else m3
    new_m4 = np.mean(cluster4, axis=0) if cluster4 else m4

    return np.array(cluster1), np.array(cluster2), np.array(cluster3), np.array(cluster4), new_m1, new_m2, new_m3, new_m4

# Run 10 iterations
for i in range(10):
    print(f"Iteration {i+1}")

    cluster1, cluster2, cluster3, cluster4, m1, m2, m3, m4 = k_means_iteration(X, m1, m2, m3, m4)

    # Results after each iteration
    print("Cluster 1 (around m1):", cluster1.shape[0], "points")
    print("Cluster 2 (around m2):", cluster2.shape[0], "points")
    print("Cluster 3 (around m3):", cluster3.shape[0], "points")
    print("Cluster 4 (around m4):", cluster4.shape[0], "points")
    print("Updated centroid m1:", m1)
    print("Updated centroid m2:", m2)
    print("Updated centroid m3:", m3)
    print("Updated centroid m4:", m4)
    print("\n")
```
