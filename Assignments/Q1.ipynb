```python
# Importing pandas for data manipulation
import pandas as pd
```

```python
# Path to the dataset
file_path = '..\Datasets\Titanic.csv'

# Reading the Titanic dataset in CSV format
data = pd.read_csv(file_path)

# Display the first few rows of the dataset
data.head()
```

```python
# Selecting specific columns: 'Name' and 'Age'
data[['Name', 'Age']].head()
```

```python
# Setting 'PassengerId' as the index of the dataset
data.set_index('PassengerId', inplace=True)

# Display the first few rows after indexing
data.head()
```

```python
# Sorting the data by the 'Age' column in ascending order
sorted_data = data.sort_values(by='Age', ascending=True)

# Display the first few rows of the sorted data
sorted_data[['Name', 'Age']].head()
```

```python
# Generating a statistical summary of the numerical columns in the dataset
data.describe()
```
