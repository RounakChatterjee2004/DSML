```python
import pandas as pd

# File path for the Telecom Churn dataset
file_path = '../Datasets/Telecom Churn.csv'

# Load the dataset into a DataFrame
data = pd.read_csv(file_path)

# Display the first few rows of the dataset
data.head()
```

```python
# Filter only numeric columns
numeric_data = data.select_dtypes(include=['number'])

# Display numeric columns
numeric_data.head()
```

```python
# Compute minimum values for numeric columns
minimum_values = numeric_data.min()

print("Minimum values for each feature:")
print(minimum_values)
```

```python
# Compute maximum values for numeric columns
maximum_values = numeric_data.max()

print("Maximum values for each feature:")
print(maximum_values)
```

```python
# Compute mean values for numeric columns
mean_values = numeric_data.mean()

print("Mean values for each feature:")
print(mean_values)
```

```python
# Compute range (max - min) for numeric columns
range_values = numeric_data.max() - numeric_data.min()

print("Range for each feature:")
print(range_values)
```

```python
# Compute standard deviation for numeric columns
std_dev = numeric_data.std()

print("Standard deviation for each feature:")
print(std_dev)
```

```python
# Compute variance for numeric columns
variance_values = numeric_data.var()

print("Variance for each feature:")
print(variance_values)
```

```python
# Compute percentiles for numeric columns (e.g., 25th, 50th, 75th)
percentiles = numeric_data.quantile([0.25, 0.50, 0.75])

print("Percentiles for each feature:")
print(percentiles)
```
