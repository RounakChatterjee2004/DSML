```python
# Importing necessary libraries
import pandas as pd
import numpy as np
```

```python
file_path = "..\Datasets\Covid Vaccine Statewise.csv"  # Adjust path as needed
covid_data = pd.read_csv(file_path)

# Display the first few rows of the dataset
covid_data.head()
```

```python
# Dataset description: Structure and basic statistics
print("Dataset Information:")
covid_data.info()

print("\nDataset Description (Numerical Features):")
print(covid_data.describe())

print("\nColumn Names:")
print(covid_data.columns)
```

```python
# Calculate the total number of males vaccinated
total_males_vaccinated = covid_data['Male (Doses Administered)'].sum()

# Display the result
print(f"\nTotal Number of Males Vaccinated: {total_males_vaccinated}")
```

```python
# Calculate the total number of females vaccinated
total_females_vaccinated = covid_data['Female (Doses Administered)'].sum()

# Display the result
print(f"\nTotal Number of Females Vaccinated: {total_females_vaccinated}")
```
