```python
import pandas as pd
df=pd.read_csv("/content/Covid Vaccine Statewise.csv")
df.head()
```

#1.Describe the Dataset

```python
# a. Describe the dataset
print("Dataset Description:")
print("\nShape of dataset:", df.shape)
print("_____________________________________________________________________")
print("\nColumns in dataset:")
print(df.columns)
print("_____________________________________________________________________")
print("\nFirst few rows:")
print(df.head())
print("_____________________________________________________________________")
print("\nDataset Info:")
print(df.info())
print("_____________________________________________________________________")
print("\nDataset Statistics:")
print(df.describe())
```

#2.Number of persons state wise vaccinated for first dose in India

```python
# First Dose Analysis - Step by Step
# 1. Remove India from states (keep only individual states)
states_data = df[df['State'] != 'India']

# 2. Group by state and get first dose data
first_dose = states_data.groupby('State')['First Dose Administered'].max()
print("\nFirst Dose Vaccination by State :")
print(first_dose)
```

#3. Number of persons state wise vaccinated for second dose in India

```python
# Second Dose Analysis - Step by Step
# 1. Remove India from states (keep only individual states)
states_data = df[df['State'] != 'India']
# 2. Group by state and get first dose data
second_dose = states_data.groupby('State')['Second Dose Administered'].max()
print("\nSecond Dose Vaccination by State :")
print(second_dose)
```

#Why Use max() while grouping the data?
Data Context:

It appears that the dataset may contain multiple entries for each state, possibly representing data from different dates or reporting periods.
max() ensures that you capture the most updated (highest) value of "First Dose Administered" for each state.
