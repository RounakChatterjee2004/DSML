```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('titanic')
```

```python
df.head()
```

```python
# Plot histogram for 'fare' column
sns.histplot(df['fare'], bins=30, kde=True)  # bins=30 for more granularity, kde=True adds the density curve
plt.title('Distribution of Ticket Prices (Fare) on Titanic')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.show()
```

#Output:
The histogram  shows how the fare is distributed among the passengers. The X-axis represents the ticket prices (fare), and the Y-axis represents the frequency (how many passengers had a fare in each price range). The KDE curve provides a smoother representation of the distribution.
