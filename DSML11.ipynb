#2.Histogram

```python
#So we can only use  numeric features to plot histogram
numeric_features = ["sepal_length", "sepal_width", "petal_length", "petal_width"]

for feature in numeric_features:
    plt.hist(df[feature], bins=5, color="skyblue",edgecolor="black")
    plt.title(f"Histogram of {feature}")
    plt.xlabel(feature)
    plt.ylabel("Frequency")
    plt.show()
```
