#1.Minimum Values

```python

print(data.min())
```

#2.Maximum Values


```python
print(data.max())
```

#3.Mean Values

```python
print(data['total day calls'].mean())
print(data['total day minutes'].mean())
print(data['total day charge'].mean())

# print(data.mean())
```

#4 Range Values


```python
#(Range represents the diference between highest values and lowest values)
#1 Range of total day mintues
print(data['total day minutes'].max()-data['total day minutes'].min())
print(data['total day charge'].max()-data['total day charge'].min())
```

#5.Standard Deviation


```python
print(data['total day minutes'].std())
print(data['total day charge'].std())
```

#6.Variance

```python
print(data['total day minutes'].var())
print(data['total day charge'].var())
```

#7.Percentiles (25th, 50th, 75th)
#The code percentiles_values = data['total day calls'].quantile([0.25, 0.50, 0.75]) is calculating the percentiles of the 'total day calls' column in the data DataFrame. The quantile function returns values at specific percentiles for the data, in this case, the 25th, 50th, and 75th percentiles.

#0.25 (25th percentile): This represents the value below which 25% of the data points fall. In your case, 87 means that 25% of the entries in 'total day calls' are less than or equal to 87.
#0.50 (50th percentile or median): This represents the middle value of the data. 101 means that half of the data points are less than or equal to 101.
#0.75 (75th percentile): This represents the value below which 75% of the data points fall. 114 means that 75% of the entries in 'total day calls' are less than or equal to 114.



```python
percentiles_values = data['total day calls'].quantile([0.25, 0.50, 0.75])
print(percentiles_values)
```
